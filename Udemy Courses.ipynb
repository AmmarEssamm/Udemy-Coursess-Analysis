## Section Four (lectures)

```python
def func_draw(x):
    plt.figure(figsize=(20,10))
    sns.barplot(x= 'level',  y=x  ,data= df,hue='subject',ci=None)
    plt.title('plot{}per level per subject'.format(x))          
    plt.show()
    # pie chart
    plt.figure(figsize=(5,5))
    df.groupby('level')[x].sum().plot(kind='pie')
    plt.title('plot of {} per level'.format(x))
    plt.show() 
```

```python
func_draw('num_lectures')
```
